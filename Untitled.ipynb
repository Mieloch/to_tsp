```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
files = glob.glob('data/*.csv')
```

```python
for f in files:
    print("#### {} ####".format(f))
    df = pd.read_csv(f)
    plt.figure(figsize=(20,10))

    x, y = df['x'].values[0:len(df)//2], df['y'].values[0:len(df)//2]
    for i in range(0, len(x)-1):
        plt.plot(x[i:i+2], y[i:i+2], 'o-', markersize = 5, color = 'r', linewidth=2)
    plt.plot([x[0], x[-1]],[y[0], y[-1]], 'o-', markersize = 5, color = 'r', linewidth=2)

    x, y = df['x'].values[len(df)//2:], df['y'].values[len(df)//2:]
    for i in range(0, len(x)-1):
        plt.plot(x[i:i+2], y[i:i+2], 'o-', markersize = 5, color = 'b', linewidth=2)
    plt.plot([x[0], x[-1]],[y[0], y[-1]], 'o-', markersize = 5, color = 'b', linewidth=2)

    for idx, row in df.iterrows():
        plt.text(row['x'], row['y'], row['id'] , fontsize=12)
    plt.savefig(os.path.join('plots', f.split('\\')[-1].split('.')[-2]), bbox_inches='tight')   
    print(' -> '.join([str(i) for i in df['id'].values[0:len(df)//2]]))
    print(' -> '.join([str(i) for i in df['id'].values[len(df)//2:]]))
```
